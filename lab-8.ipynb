```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('online_courses1.csv')

```

```python
df.head()
```

Histogram

```python
plt.hist(df['Rating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

Scatter Plot

```python
try:
  # Attempt to convert "students" to numeric (assuming it represents numbers as strings)
  df['students'] = pd.to_numeric(df['students'], errors='coerce')
  plt.scatter(df['students'], df['Rating'])
  plt.xlabel('Number of Students')
  plt.ylabel('Rating')
  plt.title('Relationship between Students and Rating')
  plt.show()
except (KeyError, TypeError) as e:
  print(f"Error encountered: {e}")
  print("Make sure 'students' column exists and contains numerical data.")
```

Line Plot

```python
plt.plot(df['students'], df['Rating'])
plt.xlabel('Students')
plt.ylabel('Rating')
plt.title('Students vs Ratings')
plt.show()
```

Bar Plot

```python
try:
 
    plt.bar(df['Rating'], df['students'])
    plt.xlabel('Rating')
    plt.ylabel('Number of Students')
    plt.title('Students Enrolled vs Rating')
    plt.show()
except (KeyError, ValueError) as e:
  print(f"Error encountered: {e}")
  print("Make sure 'course' and 'students' columns exist and contain appropriate data.")
```

Pie Plot

```python
import numpy as np

try:
  course_groups = df.groupby('course')['students'].sum() 
  proportions = course_groups / course_groups.sum() * 100  
  plt.pie(proportions, labels=course_groups.index, autopct='%1.1f%%', labeldistance=1.2, rotatelabels=45)
  plt.title('Proportion of Students per Course Category')
  plt.show()
  
except (KeyError, ValueError) as e:
  print(f"Error encountered: {e}")
  print("Make sure 'students' and 'course' columns exist and handle missing values appropriately.")
```

Area Plot


```python
# Area Plot (assuming you want to show cumulative students over time)
plt.fill_between(df['Rating'], df['students'], color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Cumulative Students')
plt.title('Cumulative Students Enrolled vs Rating')
plt.show()
```

Box Plot


```python
try:
  
    plt.bar(df['Rating'], df['students'])
    plt.xlabel('Rating')
    plt.ylabel('Number of Students')
    plt.title('Students Enrolled per Course')
    plt.show()
except (KeyError, ValueError) as e:
  print(f"Error encountered: {e}")
  print("Make sure 'Rating' and 'students' columns exist and contain appropriate data.")
```

Pair Plot

```python
import seaborn as sns

# Pair Plot (visualize relationships among multiple variables)
sns.pairplot(df[['Rating', 'Duration', 'students']])
plt.show()
```
